Módulo nativo csv

```python
import csv
archivo = open(r'hospitales2.csv', encoding='utf-8')
tabla = csv.reader(archivo)
with open('salida.csv', 'w', encoding='utf-8', newline = '') as salida:
   salida_writer = csv.writer(salida)
   next(tabla)
   salida_writer.writerow(['latitude', 'longitude', 'name', 'label'])
   for linea in tabla:
      coordenadas = linea[0][7:-1].split()
      salida_writer.writerow([coordenadas[1], coordenadas[0], linea[8], linea[3]])


```

Compresión de Listas

```python
numeros = [2,4,7,9]
cuadrados = []
for numero in numeros:
   cuadrados.append(numero ** 2)
print(cuadrados)
```

```python
cuadrados = list(map(lambda x: x ** 2, numeros))
print(cuadrados)
```

```python
cuadrados = [num ** 2 for num in numeros if num > 4]
print(cuadrados)
```

Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv(r'hospitales2.csv', encoding='utf-8')
df.head()
```

```python
coordenadas = [coord.split() for coord in df['WKT']]
coordenadas
```

```python
longitud = [float(elemento[1].strip('(')) for elemento in coordenadas]
latitud = [float(elemento[2].strip(')')) for elemento in coordenadas]
latitud
```

```python
datos = dict(latitude=latitud, longitude=longitud, name=df['DOM_NORMA'], label=df['NOM_MAP'])
df_2 = pd.DataFrame.from_dict(datos)
```

```python
df_2.head()
```

```python
pd.DataFrame(datos).to_csv('hospitales_salida.csv', index=False)
```

```python
df_3 = df.copy()
```

```python
# Otra forma

df_3[['Cod','longitude', 'latitude']] = df_3['WKT'].str.split(' ', expand=True)
```

```python
df_3 = df_3[['latitude','longitude','DOM_NORMA','NOM_MAP']]
df_3 = df_3.rename(columns = {'DOM_NORMA':'name','NOM_MAP':'label'})


df_3.head()
```

```python
df_3.latitude = df_3.latitude.str.replace(')','',regex=False).astype(float)
df_3.longitude = df_3.longitude.str.replace('(','',regex=False).astype(float)
df_3.head()
```
